### Purpose:

* The purpose of this Final Project is to explore how different climate and environmental factors contribute to the frequency of extreme weather events across countries.
* The key **dependent variable (DV)** is the number of extreme weather events (e.g., floods, storms, droughts, wildfires).
* Key **independent variables (IVs)** include:
    * CO₂ emissions (metric tons per capita)
    * Renewable energy consumption (% of total energy)
    * Forest area (% of land)
    * Average annual temperature (°C)
* The combined dataset will include approximately 20–30 countries observed during 2000–2023 (around 500–700 total records). The goal is to identify which climate factors are most strongly associated with the increasing frequency of extreme weather phenomena.

<br>

### Dataset source:

The data come from two verified open datasets that provide complementary information for this analysis.
* The first dataset comes from the [World Development Indicators (WDI)](https://datacatalog.worldbank.org/search/dataset/0037712/world-development-indicators) collection by the [World Bank](https://data.worldbank.org/).
It contains verified historical country-level data on climate and environmental indicators, including CO₂ emissions, renewable energy consumption, forest area, and precipitation.
The data are available for download [here](https://datacatalog.worldbank.org/search/dataset/0037712/world-development-indicators).
* The second dataset is the [Natural Disasters Dataset](https://ourworldindata.org/natural-disasters)
 from [Our World in Data](https://ourworldindata.org/), which compiles annual, country-level statistics on floods, storms, droughts, wildfires, and other natural disasters.
The data are derived from the EM-DAT International Disaster Database by the Centre for Research on the Epidemiology of Disasters (CRED) and are publicly available for download [here](https://github.com/owid/owid-datasets/blob/master/datasets/Natural%20disasters%20(EMDAT)/Natural%20disasters%20(EMDAT).csv).
