### Figure 2

Use `data-fig2.npy` for this figure.

```python
b = np.load( 'data-fig2.npy' )
labels = ['Red', 'Green', 'Blue']

plt.figure()
plt.plot( b[:,0], color = 'r' )
plt.plot( b[:,1], color = 'g' )
plt.plot( b[:,2], color = 'b' )
plt.xlabel("X value")
plt.ylabel("Y value")
plt.legend( labels )
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c = np.load( 'data-fig3.npy' )

plt.figure()
plt.scatter( c[:,0], c[:,1], color = 'b', alpha=0.3, s=200 )
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d = np.load( 'data-fig4.npy' )
xlabels = ['A', 'B', 'C', 'D', 'E']

plt.figure()
plt.bar( np.arange(len(d)), d, color = 'lightgrey' )
plt.xticks( ticks=np.arange(len(d)), labels=xlabels )
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
f = np.load( 'data-fig5.npy' )

plt.figure()
plt.pcolor( f, cmap='RdBu', shading='auto', vmin=-1, vmax=1 )
plt.colorbar()
plt.show()
```
