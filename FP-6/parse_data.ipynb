# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv('superstore.csv', encoding='windows-1252')

df0.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- Profit (the profit generated by each sales transaction)

**IVs**:
- Sales (total revenue from the order)
- Quantity (number of units purchased)
- Discount (the percentage discount applied to the order)


<br>
<br>

Let's extract the relevant columns:

```python

df = df0[['Profit','Sales','Quantity','Discount','Category', 'Sub-Category','Region','Segment','Ship Mode']]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'Sales': 'Sales_amount'} )

df.describe()
```
